```python
def main():
    while True:
        clear_output(wait=True)
        request = session.get(url,headers=header).json()
        df = request['advance']
        df1 = pd.DataFrame(request['data']).fillna(0)
        selected_columns = ['lastUpdateTime', 'symbol', 'lastPrice',
                            'change', 'pChange','ffmc', 'yearHigh','yearLow', 'totalTradedVolume', 
                            'totalTradedValue','nearWKH', 'nearWKL', 'perChange365d', 'perChange30d','date365dAgo',  'date30dAgo']

        display(df,df1[selected_columns])
        time.sleep(2)
```

```python
def main_run():
   try:
         clear_output(wait=True)
         main()
   except Exception as e:
         clear_output(wait=True)
         print(f'trying to retry to re-slove issue')
      #    raise e
         main_run()

main_run()
```
