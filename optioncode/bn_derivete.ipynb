```python
# from pandasql import sqldf
from IPython.display import display, HTML,clear_output
import requests
import pandas as pd
import time
# from bs4 import BeautifulSoup
```

```python
url ='https://www.nseindia.com/api/liveEquity-derivatives?index=nifty_bank_opt'
url_f ='https://www.nseindia.com/api/liveEquity-derivatives?index=nifty_bank_fut'

header = {
    "Connection": "keep-alive",
    "Cache-Control": "max-age=0",
    "DNT": "1",
    "Upgrade-Insecure-Requests": "1",
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) "
                  "Chrome/111.0.0.0 Safari/537.36",
    "Sec-Fetch-User": "?1", "Accept": "*/*", "Sec-Fetch-Site": "none", "Sec-Fetch-Mode": "navigate",
    "Accept-Encoding": "gzip, deflate, br", "Accept-Language": "en-US,en;q=0.9,hi;q=0.8"
    }
session = requests.session()
nse_live = session.get("http://nseindia.com", headers=header)
```

```python
def main():
    while True:
        clear_output(wait=True)
        request = session.get(url,headers=header).json()['data']
        request_f = session.get(url_f,headers=header).json()['data']
        df1 = pd.DataFrame(request).fillna(0)
        df_f = pd.DataFrame(request_f).fillna(0)
        max_pe = df1[df1['optionType'] == 'Put']['openInterest'].nlargest(3)
        max_ce = df1[df1['optionType'] == 'Call']['openInterest'].nlargest(3)
        max_trae_pe = df1[df1['optionType'] == 'Put']['noOfTrades'].nlargest(3)
        max_trade_ce = df1[df1['optionType'] == 'Call']['noOfTrades'].nlargest(3)

        # Filter the DataFrame based on the maximum value
        opc_df = pd.concat([df1[df1['openInterest'].isin(max_pe)],df1[df1['openInterest'].isin(max_ce)]],ignore_index=True)
        no_df = pd.concat([df1[df1['noOfTrades'].isin(max_trae_pe)],df1[df1['noOfTrades'].isin(max_trade_ce)]],ignore_index=True)
        selected_columns = ['expiryDate', 'optionType', 'strikePrice', 'lastPrice', 'change',
                            'highPrice', 'lowPrice', 'closePrice', 'volume', 'totalTurnover', 
                            'value', 'premiumTurnOver', 'underlyingValue', 'openInterest', 'noOfTrades']

        display(df_f[selected_columns],opc_df[selected_columns],no_df[selected_columns])
        time.sleep(2)
```

```python
def main_run():
   try:
         clear_output(wait=True)
         main()
   except Exception as e:
         clear_output(wait=True)
         print(f'trying to retry to re-slove issue')
      #    raise e
         main_run()

main_run()
```
